```python
# Importamos de la libreria la clase SVC, la cual es capaz de clasificar segun el algoritmo del SVM
from sklearn.svm import SVC
# SVC tiene una amplia gamma de parametros
# de los cuales nos interesan:
#  ·kernel 
#  ·C 
#  ·gamma
#
#·kernel; Es el tipo de transformacion que realizaremos a los datos de entrada, con esto lograremos crear limites
#          de decision Lineales en una dimension superior al plano de nuestros datos, logrando seprar de manera Lineal
#          nuestro set de caracteristicas en una dimension superior, lo que se traduce en limites de desicion NO LINEALES
#          para la dimension de nuestros datos
#          kernel : string, optional (default=’rbf’) -> ‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’
#
# ·C     ; Es el termino que maneja la REGULARIZACION del algoritmo, esto nos permite manejar el trade-off entre
#          una barrera de desicion suave y una barrera que concienta a todos los datos de entrenamiento (Lmite ALOCADO)
#          A valores bajos crea barreras de desicion muy suaves y a valores altos crea limites de desicion muy complejos
#          float, optional (default=1.0) -> ]0,infinito[
#
# ·gamma ; Es un parametro/coeficiente de los Kernel's ‘rbf’, ‘poly’ and ‘sigmoid’
#          A valores altos tiende a crear limites de decision muy complejos, por lo que es un buen proxy para regularizar
#          los SVM, pudiendo jugar con (gamma, C) para lograr obtener el mejor limite de decision a cada problema
#           float, optional (default=’auto’) -> ]0,infinito[

clf = SVC(kernel='rbf',gamma=1,C=10000)
clf.fit(X, Y)
```

```python
Xvalues_mesh = np.linspace(X[:,0].min(), X[:,0].max(), 200)
Yvalues_mesh = np.linspace(X[:,1].min(), X[:,1].max(), 200)

# Creamos dos mallas en 2D, cada una de estas mallas guarda el valor de X e Y
# para el punto (i,j), de esta forma X[i,j] nos entrega el valor X del punto (i,j),
# lo mismo para el punto Y[i,j], nos entrega el valor Y del punto (i,j)
# El punto (i,j) es un sistema de cordenada ajeno a X e Y, es referencial con respecto a los puntos y no a las cordenadas

X_mesh, Ymesh = np.meshgrid(Xvalues_mesh, Yvalues_mesh)

# X_mesh.ravel() extiende el arreglo [40L,40L] en uno de 1600L

# np.c_[A,B] para todo los valores de los arreglos A u B, extrae los valores
#  i-esimos del arreglo A y B, combinandolos en un nuevo arreglo [A[i],B[i]]

# np.c_[[1,2,3],[4,5,6]]
#array([[1, 4],
#       [2, 5],
#       [3, 6]])


Z = clf.predict(np.c_[X_mesh.ravel(), Ymesh.ravel()])        
Z = Z.reshape(X_mesh.shape)

plt.contour(X_mesh,
            Ymesh,
            Z,
            linestyles=['--', '-', '--'],
            levels=[-.9, 0, .9],
           linewidths=1,
           colors='BLACK',
           label='Hola')

plt.scatter(X[Y==1][:,0],
            X[Y==1][:,1],
           color='BLUE',
           s=5,
           label='Fast')
plt.scatter(X[Y==-1][:,0],
            X[Y==-1][:,1],
           color='RED',
           s=5,
           label='Slow')
plt.legend(bbox_to_anchor=(1, 1),loc=2)
plt.title('Clasificacion de la velocidad, segun la irregularidad y pendiente de callle para el plano XY ',size=15)
plt.xlabel('irregularidad')
plt.ylabel('pendiente')
plt.show()
```
